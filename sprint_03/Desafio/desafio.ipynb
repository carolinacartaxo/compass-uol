```python
# Importar bibliotecas requeridas para o desafio
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Abrir arquivo .CSV e montar daframe
df = pd.read_csv('googleplaystore.csv')
df.head()
```

```python
# Contar a quantidade de linhas e colunas do df
df.shape
```

### 1. Remoção de linhas duplicadas

```python
# Identificar valores duplicados
df.duplicated(keep=False).sum()

df[df.duplicated(keep=False)]
```

```python
# Dropar as linhas duplicadas
df = df.drop_duplicates()
df
```

```python
# Converter a coluna 'Reviews' para tipo numérico
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df['Reviews'].head()
```

```python
# Coluna 'Installs' após remover caracteres indesejados
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)
print(df['Installs'].head())
```

```python
# Converter a coluna 'Installs' para valor numérico
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
# Não usei o astype(int) porque uma das linhas tem um valor errado de nome 'Free' então vai dar erro
```

```python
print(df['Installs'].head())
```

```python
# Verificar os tipos de dados do df
df.dtypes
```

```python
# Remover valores indesejados da coluna 'Price' 
df['Price'] = df['Price'].str.replace('$', '')
```

```python
# Converter a coluna 'Price' para valor numérico
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df.dtypes
```

### Análises feitas com a base de dados

```python
# Vamos olhar os 20 aplicativos mais instalados para fazer o nosso top 5
df = df.sort_values(by='Installs', ascending=False)
df.head(20)
```

```python
# Se eu só pegar os apps mais instalados, eu consigo ver que temos aplicativos com nomes repetidos
# Como muitos aplicativos se repetem, vamos dropar as duplicatas do subset 'App'

df = df.drop_duplicates(subset=['App'])
df
```

```python
# Temos um app com uma categoria não existente que está escrita como '1.9'. 
# Vamos dropar esta linha porque grande parte dos seus dados estão errados ou como Nan (not a number)
df = df.drop([10472])
```

```python
# Agora vamos pegar só o top 5 apps mais baixados
# Para ficar melhor distribuído, visto que temos muitos apps com mais de um milhão de downloads, vou colocar o número de reviews como critério também

top_5_apps = df.sort_values(by=['Installs', 'Reviews'], ascending=False)
top_5_apps = top_5_apps[['App', 'Installs']].head(5)
top_5_apps
```

### 2. Criação do gráfico de barras dos 5 aplicativos mais instalados

```python
# Criando o gráfico de barras dos 5 aplicativos mais instalados considerando as reviews
plt.figure(figsize=(12, 4))
plt.bar(top_5_apps['App'], top_5_apps['Installs'], color='green')
plt.xlabel('Aplicativos')
plt.ylabel('Número de instalações')
plt.title('Top 5 Aplicativos mais instalados por quantidade de reviews')
plt.xticks(rotation=45)
plt.show()
```

```python
# Categorias de apps que mais aparecem
top_categorias = df['Category'].value_counts()
top_categorias
```

### 3. Gráfico de pizza mostrando as categorias de apps existentes no dataset de acordo com a frequência que elas aparecem

```python
# Gráfico de pizza mostrando as categorias de apps existentes no dataset de acordo com a frequência que elas aparecem
plt.figure(figsize=(20, 20))
plt.pie(top_categorias, labels=top_categorias.index, autopct='%1.1f%%', startangle=180)
plt.title('Distribuição de Apps por Categoria')
plt.show()
```

### 4. Aplicativo mais caro do Dataset

```python
# Mostrando o aplicativo mais caro existente no dataset
app_mais_caro = df.sort_values(by=['Price'], ascending=False).head(1)
print(f"O app mais caro é {app_mais_caro.iloc[0]['App']} com o preço de ${app_mais_caro.iloc[0]['Price']}.")
```

### 5. Contagem de apps classificados como 'Mature 17+'

```python
# Contagem de apps classificados como 'Mature 17+'
apps_mature_17 = df[df['Content Rating'] == 'Mature 17+'].shape[0]
apps_mature_17
print(f"Há {apps_mature_17} apps classificados como 'Mature 17+'.")
```

### 6. Top 10 apps por número de reviews bem como o respectivo número de reviews ordenado de forma decrescente

```python
# Top 10 apps por número de reviews bem como o respectivo número de reviews ordenado de forma decrescente
top_app_reviews = df.sort_values(by=['Reviews'], ascending=False)
show_top_app_reviews = top_app_reviews[['App', 'Reviews']].head(10)
show_top_app_reviews
```

### 7. Criação de pelo menos mais 2 cálculos sobre o dataset apresentados um em formato de lista e outro em formato de valor.

```python
# Média de avaliação dos apps

media_rating = df['Rating'].mean()

print(f"A média de avaliação dos apps é {media_rating:.2f}.")
```

```python
# Avaliação média de apps por categoria
media_rating_categoria_agrupado = df.groupby('Category')['Rating'].mean().reset_index().round(1)
media_rating_categoria = media_rating_categoria_agrupado.sort_values(by='Rating', ascending=False)
media_rating_categoria
```

### 8. Criação de gráficos de linhas e de dispersão para os exemplos criados acima

```python
# Criação do gráfico de linhas para a Média de Avaliação por Categoria
plt.figure(figsize=(18, 10))
plt.plot(media_rating_categoria['Category'], media_rating_categoria['Rating'], marker='o')
plt.xticks(rotation=65)
plt.xlabel('Categoria')
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Categoria')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de Dispersão com a correlação do número de instalações com as avaliações
plt.figure(figsize=(10, 6))
plt.scatter(df['Installs'], df['Reviews'], alpha=0.5)
plt.xlabel('Número de Instalações')
plt.ylabel('Número de Avaliações')
plt.title('Número de Instalações vs Número de Avaliações')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
